```python
import pandas as pd
import numpy as np
import requests
import yfinance as yf
from IPython.display import display
```

```python
def get_data(ticker):
    ticker = yf.Ticker(ticker).history(period="max", )
    df = pd.DataFrame(ticker)   # convert to df
    df = df.drop(columns=['Dividends', 'Stock Splits'])
    return df

bank_tickers = ['NA', 'TD.TO', 'BNS', 'BMO', 'RY.TO', 'CWB', 'CM.TO']

for bank in bank_tickers:
    print('This is the data for ticker: ', bank)
    stock = get_data(bank)
    display(stock)

```
